# 강의노트
### OpenCV 도형그리기

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 라인함수 그리기

image = np.full((512, 512, 3), 255, np.uint8)
# 시작점, 끝점, 색깔 (RGB), 선두께
image = cv2.line(image, (0,0), (255, 255), (255, 0 ,0), 5)

plt.imshow(image)
plt.show()
```

```python
# 사각형만들기 (이미지, 시작점, 끝점, RGB, 두께)

image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.rectangle(image, (20,20), (450, 450), (255, 200, 0), -1)

plt.imshow(image)
plt.show()
```

```python
# 원 그리기 (이미지, 센터, 반지름, 색, 두께(-1은 꽉채우기))

image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.circle(image, (255, 255), 50, (0, 200, 200), 5)

plt.imshow(image)
plt.show()
```

```python
# 다각형 그리기(이미지, 점들, 닫힌도형여부, 색, 두께 )

image = np.full((512, 512, 3), 255, np.uint8)
points = np.array([[90, 50], [50, 100] , [80,400], [350, 450], [425, 350]])
image = cv2.polylines(image, [points], True, (0, 200, 200), 5)

plt.imshow(image)
plt.show()
```

```python
# 텍스트 그리기(이미지, 텍스트, 위치, 글씨체, 폰트크기, 색)

image = np.full((512, 512, 3), 255, np.uint8)
image = cv2.putText(image, "Yellow", (100, 250), cv2.FONT_HERSHEY_SIMPLEX, 4, (5, 0, 0))


plt.imshow(image)
plt.show()
```
